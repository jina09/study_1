### 리스트 [ ]

#### 리스트 값 추가, 정렬, 삭제, 추출

추가 append, insert

정렬 reverse, sort

삭제 remove

추출 pop

```python
index=[10,40,30,60,50]

# 끝자리 추가
index.append(70) # 마지막에 70을 추가해줌
print(index)

# 중간에 추가
index.insert(1, 90) # 첫번째 자리에 숫자 99를 넣어줌
print(index)

# 정렬
index.reverse() # 리스트의 순서를 역으로 바꿔줌
print(index)

index.sort() # 순차적으로 정렬
print(index)

# 삭제
index.remove(10) # 10이라는 숫자를 없애줌
print(index)

index[0] = 40
print(index) # [40, 40, 50, 60, 70, 90]

index.remove(40) # 가장 첫번째에 있는 40 값만 없애줌
print(index) 

# 추출
a = index.pop(0) # 0번째에 있는 값 추출
print(a)
print(index)
```

#### 교집합, 합집합, 차집합

& , | , -

```python
set1 = set([1,2,3,4,5,6])
set2 = set([3,4,5,6,7,8])

# 교집합
print(set1 & set2)
print(set1.intersection(set2)) # 교집합 함수

# 합집합 (or)
print(set1 | set2)
print(set1.union(set2))

# 차집합
print(set1 - set2)
print(set1.difference(set2))
```

#### 집합 추가, 수정, 삭제

```python
set1.add(99)
print(set1)

set1.update([50,60,70]) # 집합은 순서에 영향 x, 순서가 마구잡이로 넣어짐
print(set1)

set1.remove(99) # 99 지우기
print(set1)
```

### random

### ex) 학번

```python
import random as rd
rd.seed(318)

# 딕셔너리 생성
school = {}

for i in range(2019, 2023):
 for j in range(9910, 9930):
   key = str(i) + str(j) # 숫자형을 문자형으로 만들어줌
   school[key] = rd.randint(50, 100) # 랜덤하게 값 넣어줌

print(school) #key : 학번, 값은 랜덤값 

print(school.values())
print(school['20199910']) # 94

# ai 수업 수강생 학번
ai = list(school.keys())[0:20] # 20199910~ 20199920까지 20명 뽑기
print(ai)

ai_sum = 0

for i in ai:
  print(ai_sum, "     전")
  print(school[i], "     점수")
  ai_sum = ai_sum + school[i]

  print(ai_sum, "     후")

print(ai_sum)# ai 수업 수강생들의 점수의 합


```

### 어려운 예제
<Species 별로 Sepal.Length의 평균값 구하기>



```python
# <Species 별로 Sepal.Length의 평균값 구하기>
# print(set(dat["Species"])) # {'setosa', 'virginica', 'versicolor'} 중복제거

sepal_avg = {} # 중복제거(집합)

for i in list(set(dat["Species"])): # ['setosa', 'virginica', 'versicolor']
  step1 = dat[dat["Species"] == i] # dat[Species] 가 setosa 일때

  sepal_length = step1["Sepal.Length"] # setosa의 Sepal.Length

  sepal_length_list = list(sepal_length) # setosa의 Sepal.Length를 리스트로 모아줌

  length_sum = 0
  for j in range(0, 50):
    length_sum = length_sum + sepal_length_list[j] # 누적합 더하기

  sepal_avg[i] = length_sum / 50 # 평균 구하기, 하나씩 딕셔너리에 집어넣어줌

print(sepal_avg)
```

### 각 요일 별 평균 남자 통행자 수를 구하시오.

```python
print(set(dat["day"]))
man_num_avg = {}


for i in list(set(dat["day"])):
  step1 = dat[dat["day"] == i]

  count = step1["num"]

  count_list = list(count)
  
  s = 0
  for j in range(0, 24):
    s = s + count_list[j]
  
  man_num_avg[i] = s / 24

print(man_num_avg)
```

### 각 시간대별 평균 여자 통행자 수를 구하세요. 

```python
print(set(dat["time"]))
woman_num_avg = {}


for i in list(set(dat["time"])):
  step1 = dat[dat["time"] == i]

  count = step1["wnum"]

  count_list = list(count)
  
  s = 0
  for j in range(0, 7):
    s = s + count_list[j]
  
  woman_num_avg[i] = s / 7

print(woman_num_avg)
```

### 각 시간대 별 전체 통행자 수 ( num + ynum + wnum = total )를 구하세요.

```python
print(set(dat["time"]))
time_avg = {}


for i in list(set(dat["time"])):
  step1 = dat[dat["time"] == i] # time이 0일때

  man_num = step1["num"] 
  man_num_list = list(man_num)

  woman_num = step1["wnum"] 
  woman_num_list = list(woman_num)
 
  young_num = step1["ynum"] 
  young_num_list = list(young_num)
 
  s1 = 0
  s2 = 0
  s3 = 0
  for j in range(0, 7):
    s1 = s1 + man_num_list[j]
    s2 = s2 + woman_num_list[j]
    s1 = s1 + young_num_list[j]
  
  time_avg[i] = s1 + s2 + s3

print(time_avg)
```

### 각 요일별 전체 통행자 수의 평균을 구하세요.

```python
print(set(dat["day"]))
print(list(set(dat["day"])))

day_avg = {}

for i in list(set(dat["day"])):
  step1 = dat[dat["day"]==i] # day가 thr 일때

  
  man_num = step1["num"]
  man_num_list = list(man_num)

  young_num = step1["ynum"]
  young_num_list = list(young_num)

  women_num = step1["wnum"]
  women_num_list = list(women_num)

  sum1 = 0
  sum2 = 0
  sum3 = 0 
  for j in range(0,24): # 168 % 7 = 24 행
    sum1 = sum1 + man_num_list[j]
    sum2 = sum2 + young_num_list[j]
    sum3 = sum3 + women_num_list[j]

  day_avg[i] = (sum1 + sum2 + sum3) / 24

print(day_avg)
```
